```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_proj(a, b):
    plt.figure()
    ax = plt.gca()
    ax.set_xlim(0, 5)
    ax.set_ylim(0, 5)
    # Plot vector
    ax.arrow(0,0, a[0], a[1],width = 0.05, color='blue')
    ax.arrow(0,0, b[0], b[1],width = 0.05, color='red')
    
    # Projection
    b_norm = np.sqrt(sum(b**2))  
    proj = (np.dot(a, b)/b_norm**2)*b
    print(proj)
    ax.arrow(0,0, proj[0],proj[1], width = 0.05, color='yellow')
    
    ax.legend(['b', 'a', 'projection'])
    plt.show()
    
```

```python
# Dot product
a = np.array([2,1])
x = np.arange(0,a[0]+1,1)
y = np.arange(0,a[-1]+1,1)

b = np.array([1,2])
plot_proj(b,a)
```
