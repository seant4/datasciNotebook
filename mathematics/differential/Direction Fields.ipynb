```python
import matplotlib.pyplot as plt
import numpy as np
import scipy
```

Direction field for: v' = 5 - .5v

```python
def fun(t,v):
    return 5 - (0.5* v)
```

```python
T, V = np.meshgrid(np.linspace(0,25, num=15), 
                   np.linspace(0,10, num=15))

dt = np.ones(T.shape)
dv = fun(T,V)
rk_sol = scipy.integrate.solve_ivp(fun, [0,25], [0.0], method='RK45', max_step = 1.0)
radau_sol = scipy.integrate.solve_ivp(fun, [0,25], [0.0], method='Radau', max_step = 1.0)
explicit_sol = scipy.integrate.solve_ivp(fun, [0,25], [0.0], method='RK23', max_step = 1.0)
bdf = scipy.integrate.solve_ivp(fun, [0,25], [0.0], method='BDF', max_step = 1.0)

plt.quiver(T, V, dt, dv, angles='xy')
plt.plot(rk_sol.t, rk_sol.y[0], 'o')
#plt.plot(radau_sol.t, radau_sol.y[0], 'o')
#plt.plot(explicit_sol.t, explicit_sol.y[0], 'o')
#plt.plot(bdf.t, bdf.y[0], 'o')
```
