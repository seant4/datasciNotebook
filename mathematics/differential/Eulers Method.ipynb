```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

```python
# Param
# f = function
# t0 = initial time
# tf = final endpoint
# n = number of steps in interval
# y0 = initial y value

def euler_method(f, t0,tf,n,y0):
    dt = (tf - t0)/n
    t = np.zeros(n+1)
    y = np.zeros(n+1)
    y = np.zeros(n+1)
    t[0] = t0
    y[0] = y0
    for i in range(1, n+1):
        t[i] = t[i-1] + dt
        y[i] = y[i-1] + dt * f(t[i-1], y[i-1])
    return t, y
```

```python
# Function to approximate
def fun(t,y):
    return 5 - (0.2 * y**2)

T, Y = np.meshgrid(np.linspace(0,10, num=15), 
                   np.linspace(0,10, num=15))
dt = np.ones(T.shape)
dy = fun(T,Y)

plt.quiver(T, Y, dt, dy, angles='xy')
t, y = euler_method(fun, 0, 10, 100, 0)
plt.plot(t,y)
```
